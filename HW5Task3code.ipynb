```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 2000
A = np.zeros((n, n))
```

```python
#сгенерируем матрицу А
A[n-1, 0] = -1
for i in range(0, n, 2):
    A[i][i] = 4
    A[i+1][i+1] = 2
    A[i][i+1] = -2
    A[i][i-1] = -2
    A[i+1][i] = -1
for i in range(0, n-2, 2):
    A[i+1][i+2] = -1
```

```python
A
```

```python
#найдём собственные значения и вектора
```

```python
eigval, eigvec = np.linalg.eig(A)
```

```python
eigval
```

```python
#возьмём модуль, потому что есть отрицательные значения, при этом они очень маленькие по модулю
eigval = np.abs(eigval)
```

```python
#строим гистрограмму собственных значений
plt.figure(figsize=(15,5))
plt.hist(eigval, bins=75)
plt.show()
```

```python
#как видим, в спектре действительно есть щель
```

```python
#для самих частот
w = np.sqrt(eigval)
plt.hist(w, bins = 75)
plt.figure(figsize=(25,5))
```

```python
#построим пространственную структуру вблизи минималной и максимальной энергии
```

```python
m = np.argmin(w)
M = np.argmax(w)
```

```python
plt.figure(figsize=(25,5))
plt.scatter(np.arange(len(eigvec[:, m])), eigvec[:, m])
plt.title("Вблизи минимальной энергии", fontsize=25)
```

```python
plt.figure(figsize=(20,5))
plt.scatter(np.arange(len(eigvec[:, M])), eigvec[:, M])
plt.title("Вблизи максимальной энергии", fontsize=25)
```

```python
plt.figure(figsize=(25,5))
plt.scatter(np.arange(len(eigvec[:, 600])), eigvec[:, 500])
plt.title("Вблизи какой-то энергии", fontsize=25)
```
